```python
def extract_from_page(url):
#     request page, create soup and get the part of the page we want
    source = requests.get(url)
    soup = BeautifulSoup(source.content, 'lxml')
    soup2 = soup.find_all("div", {"class": "quote mediumText "})

#     lists for quotes, authors and likes
    quotes = [a.find("div", class_="quoteText").text.strip().split("\n")[0] for a in soup2]
    authors = [a.find("div", class_="quoteText").find(class_="authorOrTitle").text.strip() for a in soup2]
    likes = [a.find_all("a")[-2].text.split()[0] for a in soup2]
    
#     to get the tags, there are several tags for each quote
    tags = []
    for a in soup2:
        tag = []
        otherstuff = a.find_all("a")
        for stuff in otherstuff:
            try:
                tag.append(stuff.get("href").split("tag/")[1])
            except:
                continue

        tags.append(tag)
        
    
    return quotes, authors, likes, tags
```

```python
def extract_topic(topic, base_url):
    
    url = base_url + "/" + topic
    quotes, authors, likes, tags = extract_from_page(url)
    
    data_dict = {"quotes":quotes, "authors":authors, "likes": likes, "tags":tags}
    
    for i in range(2,101):
        
        url_n = url + "?page=" + str(i)
        print("scrapping page " + url_n)
        quotes, authors, likes, tags = extract_from_page(url_n)
        
        data_dict["quotes"].extend(quotes)
        data_dict["authors"].extend(authors)
        data_dict["likes"].extend(likes)
        data_dict["tags"].extend(tags)
        
    return data_dict
```
